## Question Answering using LLM

```python
# Load the model as Llama model
llm = Llama(
    model_path=model_path_mistral, #Path to the GGUF model file
    n_ctx=4096, #Sets the context window to 4096 tokens
    n_gpu_layers=36, #Loads 36 model layers onto GPU for faster inference (set to 0 for CPU-only)
    n_batch=512 #Number of tokens processed at once
)
```

#### Response

```python
# Define function to use the loaded Mistral model to generate response to user question
# Use default values for some parameters if values are not provided in the function input
def response(query,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    model_output = llm(
      prompt=query,
      max_tokens=max_tokens,
      temperature=temperature,
      top_p=top_p,
      top_k=top_k
    )

    return model_output['choices'][0]['text']
```

## Question Answering using LLM with Prompt Engineering

```python
# Define function to use the loaded Mistral model to generate response to user question, while prefixing the user question with a system message
# Use default values for some parameters if values are not provided in the function input
def generate_resp_with_prompt_eng(user_prompt, max_tokens=128, temperature=0.2, top_p=0.95, top_k=50, repeat_penalty=1):
    # Define the system message
    system_message = "[INST]<<SYS>>Answer the following question in a clear and consise manner<</SYS>>[/INST]"

    # Formulate the prompt using system message and the user question
    prompt = f"{system_message}\n{user_prompt}"

    response = llm(prompt = prompt,
                   max_tokens=max_tokens,
                   temperature=temperature,
                   top_p=top_p,
                   top_k=top_k,
                   repeat_penalty=repeat_penalty,
                   stop=['INST'],
                   echo=False)

    return response['choices'][0]['text']
```

### Data Chunking

```python
# Split the data from the manual into chunks of text with length no more than 512 chunks
# Let's ensure overlapping of 20 chunks between consecutive chunks to that we preserve the context
text_splitter_overlap = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    encoding_name='cl100k_base',
    chunk_size=512,
    chunk_overlap=20
)
```

```python
# Load and split the data in PDF into chunks
chunks = pdfLoader.load_and_split(text_splitter_overlap)
```

### Embedding

Let's encode the blocks of chunks into vector representations. We will use the "gte-large" embedding model to generate the embeddings. The embedding vector size for this model is 512, which is same as the token size of our chunking strategy

```python
# Download and load the "gte-large" embedding model from HuggingFace
embedding_model = SentenceTransformerEmbeddings(model_name='thenlper/gte-large')
```

### Vector Database

Let's generate the embeddings and store them in an output directory

```python
# Create the output directory if it doesn't exist
out_dir = 'medical_diag_db'

if not os.path.exists(out_dir):
  os.makedirs(out_dir)
```

```python
# Generate the embeddings and store them in the output directory
vectorstore = Chroma.from_documents(
    chunks,
    embedding_model,
    persist_directory=out_dir
)
```

Even though an error is observed in the console above, there is no issue in creating the vector database, as we will observe in the next section

### Retriever

```python
# Define the retriever to fetch documents based on similar embeddings, using cosine similarity, and return top 5 matches
retriever = vectorstore.as_retriever(
    search_type='similarity',
    search_kwargs={'k': 5}
)
```

```python
# Test the retriever on a sample query
relevant_chunks = retriever.invoke("What is the protocol for managing sepsis in a critical care unit?")
relevant_chunks
```

We are able to retreive the relevant chunks for the given input query

### System and User Prompt Template

Let's define the below 2 parts to create a template for the prompt for the RAG based system:
* System Message: Instructions to be given to the LLM
* User message template: Message template that contains the context derived from the document chunks, and the user query

```python
# Define the system message template
qna_system_message = """
You are a medical assistant whose work is to give answers to questions with respect to a context.
User input will have the context for you to answer user questions.

The context will begin with the token: ###Context.
The context contains references to specific portions of a document relevant to the user query.

User questions will begin with the token: ###Question.

Use the context provided to find accurate and reliable answers.
If the answer is not found in the context, reply with "I don't know".
Do not mention the about the provided context in your final answer.

"""
```

```python
# Define the user message template
qna_user_message_template = """
Conider the following ###Context and ###Question
###Context
{context}

###Question
{question}
"""
```

### Response Function

```python
# Define a funtion that takes the user question as an input, along with some model parameters to optimize the response
# Function will use the system message and user message templates and use the LLM to evaluate the answer to the user question
def generate_rag_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.invoke(input=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    #prompt = qna_system_message + '\n' + user_message
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""
    # Generate the response
    #print(prompt)

    try:
        response = llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k,
                  stop=['INST'],
                  echo=False
                  )

        # Extract and print the model's response
        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

Let's define the below 3 parts to create a templates for output evaluation
* Groundedness rater System Message: Instructions to be given to the LLM to evaluate the groundedness
* Relevance rater System Message: Instructions to be given to the LLM to evaluate the relevance
* User message template: Message template that contains the context derived from the document chunks, and the user query and the answer to the user query based


```python
# Define the system message for groundedness rating
groundedness_rater_system_message  = """
You will be provided a question and context used by an AI system and the answer generated by the AI system

Your task is to evaluate the extent to which the answer has been derived from the context.
Use below criteria to evaluate:
Rate it 1 - if the answer is not derived from the context at all
Rate it 2 - if the answer is derived from the context only to a limited extent
Rate it 3 - if the answer is derived from the context to a good extent
Rate it 4 - if the answer is derived from the context for most of the part
Rate it 5 - if the answer is derived from the context completely

Only give a single rating along with a brief explanation for the rating

"""
```

```python
# Define the system message for relevance rating
relevance_rater_system_message = """
You will be provided a question and context used by an AI system and the answer generated by the AI system

Your task is to evaluate the extent to which the answer is relevant to the question while considering if it directly answers the key aspect of the question based on the context provided.
Use below criteria to evaluate:
Rate it 1 - if the answer is not relevant to the question at all
Rate it 2 - if the answer is only slightly related to the question and missing key aspects
Rate it 3 - if the answer is moderately related to the question and addresses some parts of the question but leaves out important details
Rate it 4 - if the answer is mostly relevant and addresses most parts of the question
Rate it 5 - if the answer is fully relevant and addresses all aspects of the question with details from the context

Only give a single rating along with a brief explanation for the rating

"""
```

```python
# Define the user message template
user_message_template = """
###Question
{question}

###Context
{context}

###Answer
{answer}
"""
```

```python
# Define a funtion that takes the user question as an input, along with some model parameters to optimize the response
# Function will use the system messages for the ratings and user message templates and use the LLM to evaluate the ratings for groundedness and relevance
def generate_ground_relevance_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.invoke(input=user_input,k=3)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Evaluate groundedness of the answer
    response_1 = llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    # Evaluate relevance of the answer
    response_2 = llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```
